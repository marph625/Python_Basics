#### **Lv 18 Quest 3 Drawing Different Shapes**
Triangle  |  Square  |  Pentagon  |  Hexagon  |  Heptagon  |  Octagon  |  Nonagon  |  Decagon

- bei jeder gezeichneten Linie soll eine zufllige Farbe aus der color_list ausgewhlt werden



```python
import random

color_list = ["aquamarine", "aquamarine4", "bisque4", "blue", "BlueViolet", "brown1",
            "CadetBlue1", "chartreuse", "chartreuse4", "chocolate", "coral2",
            "DarkOliveGreen3", "DarkGoldenrod1", "DarkGreen", "DarkOrange1", "DarkRed",
            "DarkSeaGreen3", "DarkSlateGray", "DeepPink2", "DarkTurquoise", "DeepSkyBlue"]
corners = 3
sum_angles = 360
turtle_movespeed = 80

while corners <= 10:
    for _ in range(corners):
        myDudu.fd(turtle_movespeed)
        myDudu.left(sum_angles / corners)
        myDudu.pencolor(random.choice(color_list))
    corners += 1
```
