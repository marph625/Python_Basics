## **Constructor**

\_\_init\_\_ Funktion wird jedes mal aufgerufen, wenn ein neues Objekt einer Klasse erzeugt wird

```python
class Tea:
    def __init__(self):
        print("New tea being created...")
        
tea_1 = Tea

# 1. Markiere z.B. seats, 2. STRG + D -> je öfter, desto mehr Stellen, an denen seats steht, sind gleichzeitig editierbar  

class Car:
    def __init__(self, seats, color, maxspeed):
        self.seats = seats
        self.color = color
        self.maxspeed = maxspeed

new_car = Car(5, "cherryred", 200)

new_car.seats = 3

print(new_car.seats, new_car.color, new_car.maxspeed)
```

### **User class with Constructor**

```python
class User:
    def __init__(self, user_id, username) -> None:
        self.user_id = user_id
        self.username = username
        self.followers = 0
        
user1 = User("007", "James")
user2 = User("001", "Ryu")

print(user1.username,"followers:",user1.followers)
print(user2.username,"followers:",user2.followers)
```

### **Adding some Methods**

```python
class User:
    def __init__(self, user_id, username) -> None:
        self.user_id = user_id
        self.username = username
        self.followers = 0
        self.following = 0
        
    def follow(self, user):
        user.followers += 1
        self.following += 1
        
user1 = User("001", "Supermario")
user2 = User("002", "Link")
user3 = User("003", "Sonic the Hedgehog")

user1.follow(user2)

print(f"{user1.username} Followers: {user1.followers} Following: {user1.following}")
print(f"{user2.username} Followers: {user2.followers} Following: {user2.following}")
print(f"{user3.username} Followers: {user3.followers} Following: {user3.following}")
```
