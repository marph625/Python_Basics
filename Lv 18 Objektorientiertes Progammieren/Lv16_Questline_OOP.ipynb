### **Warum OOP?**

+ Code wird immer komplexer
+ Zunahme von Abhängigkeiten und Beziehungen
+ Prozedurale Imperative Programmierung
+ Eine Person im Restaurant, der aufräumt, kocht, Bestellungen entgegen nimmt

***

***

### **Was ist OOP?**

+ Aufbrechen in einzelne Module
+ Aufteilung in einzelne Aufgaben
+ Beispiel: Ein Elektroauto mit entsprechenden Systemen
+ Objekte haben etwas und können etwas machen

***

### **Wie ?**

+ Von einem Objekt können mehrere Versionen erstellt werden
+ dazu muss eine Blaupause für die Objekte erzeugt werden -> mehrere Kellner
+ diese Blaupausen oder Baupläne werden Klassen genannt
